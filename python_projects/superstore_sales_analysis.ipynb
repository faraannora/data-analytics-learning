## Import file dari google drive, saat dibuka sebagai file dengan url

```python
# membuat variabel url ('edit?pli=1&gid=1961927276#' diganti 'export?format=csv&'), lalu baris kedua membuat variabel untuk membaca url

url="https://docs.google.com/spreadsheets/d/1x_nOZV-LrPhCasIMQtPvu6Rd_jkRIXFxEocP53hbrQk/export?format=csv"


data_url = pd.read_csv(url)
```

## Analisis Deskriptif

```python
### cek tipe data dan missing value
data_url.info()
```

```python
### menampilkan informasi statistik
data_url.describe().round(3)
```

##DATA CLEANING

```python
### mengubah kolom tanggal dair tect menjadi tipe tanggal

data_url['Order Date']=pd.to_datetime(data_url['Order Date'], dayfirst=True)
```

```python
data_url.info()
```

```python
data_url['Discount'] = pd.to_numeric(data_url['Discount'], errors='coerce')
data_url['Profit'] = pd.to_numeric(data_url['Profit'], errors='coerce')
```

```python
# korelasi

corr = data_url[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
print(corr)
```

```python
# melihat trend
sales_trend = data_url.groupby('Order Date')['Sales'].sum()
```

```python
sales_trend.plot(figsize=(10,5), title='Sales Trend Over Time')
```

```python
import matplotlib.pyplot as plt

monthly_sales = data_url.resample('M', on='Order Date')['Sales'].sum()

monthly_sales.plot(figsize=(10,5), title='Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```
