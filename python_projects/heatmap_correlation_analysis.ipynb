```python
import pandas as pd
```

```python
country_data=pd.read_excel('country_data.xlsx')
country_data
```

```python
country_data.info()
```

```python
corr=country_data.drop(['country'],axis=1).corr()
corr
```

```python
#melihat trend
import matplotlib.pyplot as plt
scatter=plt.scatter(country_data['income'], country_data['gdpp'])
plt.xlabel('income')
plt.show('gdpp')
```

```python
import matplotlib.pyplot as plt
scatter=plt.scatter(country_data['income']/1_000_000,
            country_data['gdpp']/1_000_000)
plt.title('Income vs GDP per Capita')
plt.xlabel('income (million)')
plt.ylabel('gdpp per capita(million)')
plt.show()
```

```python
import seaborn as sns

sns.scatterplot(x='income', y='gdpp', data=country_data)
```

```python
import seaborn as sns

sns.scatterplot(
    x=country_data['income']/1_000_000,
    y=country_data['gdpp']/1_000_000
)

plt.xlabel('Income (Million)')
plt.ylabel('GDP per Capita (Million)')
plt.show()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
